* 객체는 값, 유형, 정체성으로 구성되어 있다

```python
year=1789
print(year,
      type(year),
      id(year))

new_year=1789
print(year == new_year,              # 두 객체의 값 비교
      type(year) == type(new_year),  # 두 객체의 유형 비교
      id(year) == id(new_year))      # 두 객체의 정체성 비교

same_year=year
print(year == same_year,              # 두 객체의 값 비교
      type(year) == type(same_year),  # 두 객체의 유형 비교
      id(year) == id(same_year))      # 두 객체의 정체성 비교
```

* 일반적으로 클래스의 속성은 class 문에서 확정하고 수정하지 않는다

이름공간은 프로그래밍 언어에서 이름이 가리키는 대상을 제한하는 범위다

image.png

```python
class Cake:
    """케익을 나타내는 클래스"""
    coat = '생크림'
cake_1 = Cake()

print('1.',
      Cake.coat,
      cake_1.coat)

Cake.coat = '초콜릿'
print('2.',
      Cake.coat,
      cake_1.coat)

cake_1.coat = '아이스크림'
print('3.',
      Cake.coat,
      cake_1.coat)
```

* dir() 함수를 사용하여 이름공간(클래스)에 정의된 모든 이름(속성)의 리스트를 구할 수 있다.

####\* 초기화
인스턴트 초기 속성 정의

```python
class Cake:
    """케익을 나타내는 클래스"""
    coat = '생크림'
    
    def __init__(self, topping, price, candles=0):
        """인스턴스를 초기화한다."""
        self.topping = topping   # 케익에 올린 토핑
        self.price = price      # 케익의 가격
        self.candles = candles  # 케익에 꽂은 초 개수)

cake_2 = Cake('딸기', 50000, 25)
print(cake_2.topping,
      cake_2.price,
      cake_2.candles)
```

* __init__() 메서드는 첫번째 매개변수(self)에 인스턴스를 전달받고, 두번째 매개변수부터 사용자가 전달한 인자를 입력받아 초기화에 사용한다

###③ 매서드
데이터 유형을 다루는 방법이 정의된 함수

```python
class Cake:
    """케익을 나타내는 클래스"""
    coat = '생크림'
    
    def describe():
        """이 케익에 관한 정보를 화면에 출력한다."""
        print('이 케익은', Cake.coat, '으로 덮여 있다.')  # 클래스 속성으로 정의된 메서드

Cake.describe()

cake_1 = Cake()
cake_1.describe()
```

```python
class Cake:
    """케익을 나타내는 클래스"""
    coat = '생크림'
    
    def describe(self):
        """이 케익에 관한 정보를 화면에 출력한다."""
        print('이 케익은', self.coat, '으로 덮여 있다.')  # 인스턴스 속성으로 정의된 메서드

cake_1 = Cake()
cake_1.describe()

Cake.describe()
```

* __인스턴스를 기준으로 메서드를 호출__하면, 인스턴스에서 첫번째 인자로 인스턴스를 전달한다.  
__클래스 속성으로 정의된 메서드__는 전달되는 데이터는 있지만 이를 전달받을 매개변수가 없어 오류가 발생한다
* __클래스를 기준으로 메서드를 호출__하면, __인스턴스 속성으로 정의된 메서드__는 전달받은 인스턴스가 없어 오류가 발생한다
