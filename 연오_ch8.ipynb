```python
year=1789
print(year,
      type(year),
      id(year))

new_year=1789
print(year == new_year,              # 두 객체의 값 비교
      type(year) == type(new_year),  # 두 객체의 유형 비교
      id(year) == id(new_year))      # 두 객체의 정체성 비교

same_year=year
print(year == same_year,              # 두 객체의 값 비교
      type(year) == type(same_year),  # 두 객체의 유형 비교
      id(year) == id(same_year))      # 두 객체의 정체성 비교
```

또한 상위 클래스에 정의된 속성은 하위 클래스에서 고쳐 정의할 수 있다

```python
class Cake:
    """케익을 나타내는 클래스"""
    coat = '생크림'
    
    def __init__(self, topping, price, candles=0):
        """인스턴스를 초기화한다."""
        self.topping = topping
        self.price = price
        self.candles = candles

    def describe(self):
        print(self.coat,
              self.topping,
              self.price,
              self.candles)

class IceCreamCake(Cake):
    """아이스크림 케익을 나타내는 클래스."""
    coat = '아이스크림'  # 하위 클래스 속성 재정의하기
    cacao_percent = 32.0  # 하위 클래스만의 속성
    
    def __init__(self, flavor, topping, price, candles=0):  # __init__() 메서드 재정의하기
        """인스턴스를 초기화한다."""
        self.flavor = flavor
        super().__init__(topping, price, candles)

ice_cream_cake_1 = IceCreamCake('바닐라맛', '쿠키인형', 42000)
ice_cream_cake_1.describe()
```

* __<super()>__ 하위 클래스에서 상위 클래스를 가리킨다

```python
issubclass(IceCreamCake, Cake)
```

####②* 초기화
인스턴트 초기 속성 정의

```python
class Cake:
    """케익을 나타내는 클래스"""
    coat = '생크림'
    
    def __init__(self, topping, price, candles=0):
        """인스턴스를 초기화한다."""
        self.topping = topping   # 케익에 올린 토핑
        self.price = price      # 케익의 가격
        self.candles = candles  # 케익에 꽂은 초 개수

cake_2 = Cake('딸기', 50000, 25)
print(cake_2.coat,
      cake_2.topping,
      cake_2.price,
      cake_2.candles)
```

* __init__() 메서드는 첫번째 매개변수(self)에 인스턴스를 전달받고, 두번째 매개변수부터 사용자가 전달한 인자를 입력받아 초기화에 사용한다

###③ 매서드
데이터 유형을 다루는 방법이 정의된 함수

```python
class Cake:
    """케익을 나타내는 클래스"""
    coat = '생크림'
    
    def describe():
        """이 케익에 관한 정보를 화면에 출력한다."""
        print('이 케익은', Cake.coat, '으로 덮여 있다.')  # 클래스 속성으로 정의된 메서드

Cake.describe()

cake_1 = Cake()
cake_1.describe()
```

```python
class Cake:
    """케익을 나타내는 클래스"""
    coat = '생크림'
    
    def describe(self):
        """이 케익에 관한 정보를 화면에 출력한다."""
        print('이 케익은', self.coat, '으로 덮여 있다.')  # 인스턴스 속성으로 정의된 메서드

cake_1 = Cake()
cake_1.describe()

Cake.describe()
```

* __인스턴스를 기준으로 메서드를 호출__하면, 인스턴스에서 첫번째 인자로 인스턴스를 전달한다(인스턴스 전용 속성에 접근하기 위함)  
__클래스 속성으로 정의된 메서드__는 전달되는 데이터는 있지만 이를 전달받을 매개변수가 없어 오류가 발생한다
* __클래스를 기준으로 메서드를 호출__하면, __인스턴스 속성으로 정의된 메서드__는 전달받은 인스턴스가 없어 오류가 발생한다
