#####\* 리스트 연산

```python
number_list = [1,2,3,4,5]
alphabet_list = ['a','b','c','d','e']
list_list = [['a','b'],['c','d','e']]
print('1. +, *\n  ', number_list+alphabet_list,
      number_list*2)
print('2. 인덱싱\n  ', alphabet_list[1],
      alphabet_list[-1],
      alphabet_list[0],
      alphabet_list[-0])
print('3. 슬라이싱\n  ', alphabet_list[1:3],
      alphabet_list[::2],
      alphabet_list[1::2],
      alphabet_list[::-1])
print('4. 중첩 인덱싱\n  ', list_list[1][1])
print('5. in, not in\n  ', 3 in number_list,
      0 not in number_list)
```
