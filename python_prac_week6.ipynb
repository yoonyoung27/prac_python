```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f2(x, w):
    return (x-w)*(x+1)*x*(x-1)

x=np.linspace(-4, 4, 200)
plt.plot(x, f2(x,2), color='black', label='$w=2$')
plt.plot(x, f2(x,-2), color='cornflowerblue', label='$w=-2$')
plt.legend(loc='upper left')
plt.ylim(-5,35)
plt.title('$f_2(x)')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f3(x0, x1):
    r=2*x0**2+x1**2
    ans=r*np.sin(-r)
    return ans

xn=50
x0=np.linspace(-10, 10, xn)
x1=np.linspace(-10, 10, xn)
y=np.zeros((len(x0), len(x1)))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0]=f3(x0[i0], x1[i1])

from mpl_toolkits.mplot3d import Axes3D

xx0, xx1 = np.meshgrid(x0, x1)

plt.figure(figsize=(5, 3.5))
ax=plt.subplot(1,1,1,projection='3d')
ax.plot_surface(xx0, xx1, y, rstride=1, cstride=1, alpha=0.3, color='blue', edgecolor='black')
ax.set_zticks((0,0.2))
ax.view_init(75, -95)
plt.show()
```

```python
import math

def target_random_number(min_value=1, max_value=32, seed=2016):

    random.seed(seed)
    answer = random.randint(min_value, max_value)

    for i in range(6):
        my_answer = math.floor((min_value + max_value)/2)
        print("내가 제출한 정답은 {}입니다.".format(my_answer))

        if answer > my_answer:
            print("Up! 더 큰 숫자를 제출하세요.")
            min_value = my_answer

        elif answer < my_answer:
            print("Down! 더 작은 숫자를 제출하세요.")
            max_value = my_answer

        else:
            print("정답을 맞췄습니다. 정답은 {}이었습니다.".format(answer))
            break
        
        print('\n')

    if answer != my_answer:
        print("정답을 맞추지 못했습니다. 정답은 {}이었습니다.".format(answer))

target_random_number()      
```

```python
import math as m

def find_divisor(num):
    divisor_list = [1]
    for i in range(2, int(m.sqrt(num)+1)):
        if num%i == 0:
            divisor_list = divisor_list + [i, int(num/i)]
    return divisor_list

def is_perfect_num(num, func = find_divisor):
    if num == sum(func(num)):
        return True
    else:
        return False

test_list = [4,6,28,496]

for i in test_list:
    print(str(i) + ': ' + str(is_perfect_num(i)))
```

```python
def multiplus(n):
    num=str(n)
    sum=0
    for i in range(len(num)):
        sum+=int(num[i])**2
    return sum

def cyclic(n):
    a=[n]
    while((multiplus(a[len(a)-1])!=1) and (multiplus(a[len(a)-1])!=4)):
        a.append(multiplus(a[len(a)-1]))
    return (multiplus(a[len(a)-1]))

def happynumber(n):
    if(cyclic(n)==1):
        return True
    else:
        return False
    
for i in range(1, 1001):
    if(happynumber(i)==True):
        print(i, end=' ')
```

```python
class Turtle:
    def __init__(self):
        self.x, self.y, self.direction, self.speed = 0, 0, 0, 1

    def right(self):
        self.x += self.speed
        print('$ e')

    def left(self):
        self.x -= self.speed
        print('$ n')
    
    def down(self):
        self.y -= self.speed
        print('$ s')

    def acceleration(self):
        self.speed += 1
        if self.speed > 3:
            self.speed = 3
        print('스피드 : %d' % self.speed)

    def brake(self):
        self.speed -= 1
        if self.speed < 0:
            self.speed = 0
        print('스피드 : %d' % self.speed)
    
    def printposition(self):
        print('지금 위치는 (%d,%d)입니다.' %(self.x, self.y))

    def quit(self):
        print('$ f')
        print('(%d,%d)에서 끝납니다.' %(self.x, self.y))

jun = Turtle()

jun.acceleration()
jun.acceleration()
jun.acceleration()

jun.brake()
jun.brake()
jun.brake()
jun.brake()

jun.acceleration()

jun.printposition()

jun.down()
jun.printposition()
jun.acceleration()
jun.left()
jun.quit()
```
