###(개념) 객체

image.png

* 객체는 값, 유형, 정체성으로 구성되어 있다
* 클래스는 데이터 유형을 나타내는 데이터로, int, float, str 등이 있다

```python
year=1789
print(year,
      type(year),
      id(year))

new_year=1789
print(year == new_year,              # 두 객체의 값 비교
      type(year) == type(new_year),  # 두 객체의 유형 비교
      id(year) == id(new_year))      # 두 객체의 정체성 비교

same_year=year
print(year == same_year,              # 두 객체의 값 비교
      type(year) == type(same_year),  # 두 객체의 유형 비교
      id(year) == id(same_year))      # 두 객체의 정체성 비교
```

###① 인스턴스 만들기
인스턴스 : 어떤 클래스에 속하는 객체

* 리터럴
* 식
* 인스턴스화(클래스이름())

```python
1789      # 정수 리터럴
3.1415    # 실수 리터럴
1-2j      # 복소수 리터럴
'파이썬'  # 문자열 리터럴

5 * 5          # 산술 식 : 수 인스턴스 만들기
[1,2,3]        # 리스트 식 : 리스트 인스턴스 만들기
{'year':1789}  # 사전 식 : 사전 인스턴스 만들기
lamda x: x*x   # 람다 식 : 함수 인스턴스 만들기

int()           # 정수의 기본 인스턴스 만들기
int(1789)       # 1789에 대응하는 정수의 인스턴스 만들기
int('1789')     # '1789'에 대응하는 정수의 인스턴스 만들기
int(1789.0)     # 1789.0에 대응하는 정수의 인스턴스 만들기
list()          # 리스트의 기본 인스턴스 만들기
list((0,1,2))   # (0,1,2)에 대응하는 리스트의 인스턴스 만들기
list(range(3))  # range(3)에 대응하는 리스트의 인스턴스 만들기
```

* __<int()>__ 정수 인스턴스화
* __<float()>__ 실수 인스턴스화
* __<complex()>__ 복소수 인스턴스화
* __<str()>__ 문자열 인스턴스화
* __<bool()>__ 불리언 인스턴스화
* __<list()>__ 리스트 인스턴스화
* __<tuple()>__ 튜플 인스턴스화
* __<range()>__ 레인지 인스턴스화(ch5 참고)
* __<dict()>__ 사전 인스턴스화
* __<set()>__ 집합 인스턴스화
* __<frozenset()>__ 내용을 수정하는 연산,함수를 지원하지 않는 집합 인스턴스화


```python
type(1789)

isinstance(1789, int)
isinstance(1789, object)
```

* isinstance() 함수로 객체가 어떤 클래스의 인스턴스인지 확인 할 수 있다
* 객체 하나의 유형은 클래스 하나로 정해져 있기 때문에, type() 함수는 하나의 값만을 반환한다
* 하지만 클래스에는 계층이 있기 때문에, 한 객체는 동시에 여러 클래스의 인스턴스일 수 있다
