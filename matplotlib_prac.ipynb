###2차원 그래프

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x=np.arange(-3, 3.5, 0.01)
#x=np.linspace(-3, 3.49, 650) ; 위와 같은 식
plt.plot(x, (x-2)*x*(x+2), color='black', label='$w=2$')  # 반환
plt.legend(loc='upper left')
plt.ylim(-15, 15)
plt.title('$y=x^3-4x$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid(True)
plt.show()                                                # 출력
```

* __<$>__ latex의 수식 기호

```python
import matplotlib
matplotlib.colors.cnames
```

* color는 이름 또는 (Red, Green, Blue)의 조합으로 나타낼 수 있다 [ex. 'red' 또는 (255, 0, 0)]

```python
def f(x, w):
    return (x-w)*x*(x+2)

plt.figure(figsize=(10,3))                   # 그래프의 사이즈
plt.subplots_adjust(wspace=0.5, hspace=1.5)  # 그래프 사이의 간격
for i in range(6):
    plt.subplot(2, 3, i+1)                   # 그래프의 위치 (2행 3열의 i+1번 째 자리)
    plt.title(i+1)
    plt.plot(x, f(x,i), 'b')
    plt.ylim(-20, 20)
    plt.grid(True)

plt.show()
```

###3차원 그래프

```python
def f(x0, x1):
    r = 2*x0**2+x1**2
    ans = r*np.exp(-r)
    return ans

xn=9
x0=np.linspace(-2, 2, xn)
x1=np.linspace(-2, 2, xn)
y=np.zeros((len(x0), len(x1)))  # 9행 9열의 행렬
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f(x0[i0], x1[i1])

print(np.round(y, 1))
```

```python
plt.figure(figsize=(3.5, 3))
plt.bone()     # 그라데이션 [cf. plt.pink(), plt.gray]
plt.pcolor(y)  # y값에 따라 색을 지정
plt.colorbar()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

xx0, xx1 = np.meshgrid(x0, x1)                                                                  # 좌표를 튜플 형태로 지정
plt.figure(figsize=(5, 3.5))
ax=plt.subplot(1, 1, 1, projection='3d')
ax.plot_surface(xx0, xx1, y, rstride=1, cstride=1, alpha=0.3, color='blue', edgecolor='black')  # stride로 선 개수 지정 (수가 작을수록 선이 많아진다), alpha로 투명도 지정
ax.set_zticks((0, 0.2))
ax.view_init(75, -95)                                                                           # 그래프를 위로 75도, 반시계방향으로 95도 회전
plt.show()
```

```python
xn=50
x0=np.linspace(-2, 2, xn)
x1=np.linspace(-2, 2, xn)
y=np.zeros((len(x0), len(x1)))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0]=f(x0[i0], x1[i1])
xx0, xx1 = np.meshgrid(x0, x1)

plt.figure(1, figsize=(4, 4))
cont=plt.contour(xx0, xx1, y, 5, colors='cornflowerblue')
cont.clabel(fmt='%3.2f', fontsize=8)
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)
plt.show()
```
