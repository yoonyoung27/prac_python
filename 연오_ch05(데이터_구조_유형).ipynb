#####\* 리스트 연산

```python
number_list = [1,2,3,4,5]
alphabet_list = ['a','b','c','d','e']
list_list = [['a','b'],['c','d','e']]

print('1. +, *\n  ', number_list+alphabet_list,
      number_list*2, '\n')

print('2. 인덱싱\n  ', alphabet_list[1], '\n')

print('3. 슬라이싱\n  ', alphabet_list[1:3], '\n')

print('4. in, not in\n  ', 3 in number_list,
      0 not in number_list)
```

* __a:b__ a이상 b미만

```python
print('1. 인덱싱+\n  ', alphabet_list[-1],
      alphabet_list[0],
      alphabet_list[-0],
      list_list[1][1], '\n')

print('2. 슬라이싱+\n  ', alphabet_list[::2],
      alphabet_list[1::2],
      alphabet_list[::-1], '\n')

number_list[1:2] = [200, 300]
print('3. 수정\n  ', number_list, '\n')

del number_list[3]
print('4. 삭제\n  ', number_list, '\n')
```

* __a::b__ a번째부터 b간격으로 슬라이싱

```python
copied_list_1 = alphabet_list[:]
copied_list_1[0] = 'A'
print(copied_list_1 == alphabet_list)

copied_list_2 = alphabet_list
copied_list_2[0] = 'A'
print(copied_list_2 == alphabet_list)
```

* copied_list_1은 다른 변수에 시퀀스를 복제한 것이고, copied_list_2는 같은 변수에 이름이 하나 더 생긴 것이다

#####* 통계 함수

* __<sum()>__ 모든 요소의 합을 구한다
* __<min()>__ 시퀀스에 포함된 최소 요소를 구한다
* __<max()>__ 시퀀스에 포함된 최대 요소를 구한다

```python
numbers = [30,20,40,10]

print('1.', len(numbers),
      numbers.count(20))

numbers.sort()
print('2.', numbers)

numbers.reverse()
print('3.', numbers)

numbers.append(100)
print('4.', numbers)

numbers.insert(4,50)
print('5.', numbers)

numbers.extend([101,102,103])
print('6.', numbers)

numbers += [104,105]
print('7.', numbers)

print('8.', numbers.pop(),
      numbers.pop(1))

numbers.remove(40)
print('9.', numbers)

numbers = numbers.clear()
print('10.', numbers)
```

* extend() 메서드는 += 연산으로 대신할 수 있다
* a.sort() %>% a.reverse는 a.sort(reverse=True)와 같다
* pop(x) 메서드는 x번째 요소를 꺼낸다

```python
numbers = [30,20,40,10]

print(sum(numbers),
      min(numbers),
      max(numbers))
```

```python
print(min(alphabet_list),
      max(alphabet_list))
print(min([1,2,3,'a']))
```

* sort(), min(), max() 함수를 사용하려면 시퀀스에 포함된 요소는 서로 비교할 수 있는 대상이어야 한다
